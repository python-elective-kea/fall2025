## \__init__()

When you initialize and object, there is a corosponding \__init__() method that is called

```python
class Animal:
    def __init__(self, name):
        self.name = name

dog = Animal('Claus')
dog.name
```

## \__repr__()

If you want a string representation of the object (the objects state), you call the top-level function *repr()* which has a corosponding implementation of \__repr__

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def __repr__(self):
        return f'{self.name}'
    
    def __str__(self):
        return f'Mit object har denne attribut {self.name}'

dog = Animal('Claus')
print(dog)
```

A list in python can be accessed like with this syntax:

This is because a list implements the method

```python
def __getitem__(self):
    pass
```

Your own object, if implementing this method can have the same behaviour

```python
'''
__len__ metoden
__add__ metoden
__repr__ metoden
__str__ metoden
__setitem__ metoden
__delitem__ metoden


'''


class Deck:
    def __init__(self):
        self.cards = ['A', 'K', 4, 7]

    def __repr__(self):
        return f'{self.__dict__}'
    
    def __getitem__(self, key):
        return self.cards[key]
    
    def __setitem__(self, key, value):
        self.cards[key] = value
        return None
    
    def __delitem__(self, key):
        del(self.cards[key])
        return None
    
    def __len__(self):
        return len(self.cards)
    
    def __add__(self, other):
        newdeck = Deck()
        newdeck.cards = self.cards + other.cards
        return newdeck
    


deck = Deck()
deck2 = Deck()

4 in deck
```
