# Coroutines



```python
class Api:
    def do_this_first(self):
        return 'Jeg er den første'
    def do_this_second(self):
        return 'Jeg er den anden'
    def do_this_third(self):
        return 'Jeg er den tredje'
    
    def __iter__(self):
        pass

    def __next__(self):
        pass
    
    

api = Api()

api.do_this_second()
api.do_this_first()

api.__iter__()

iter(api)
next(api)
```

```python
def api():
    yield 'do_this_first'
    yield 'do_this_second'
    yield 'do_this_third'


# __iter__
# __next__

for i in api():
    print(i)


```

### .send() and .close()


```python
def simple_coroutine():
    print('-> coroutine started')
    x = yield 'do this first and wait for input'
    while x:
        x = yield f'this is yeilded : {x}'
```

```python
g = simple_coroutine()
next(g)
```

```python
g.send('Hello')
```

```python
g.send('slfhslkdjhfskjhfsdkjhdkfs')
```

```python
def averager(*args):
    sum = 0
    for i in args:
        sum += i
    return sum / len(args)


averager(1, 2, 3, 4, 5, 6, 7, 8)
```

```python
"""
f = open('testfiles/bohr.txt', 'r')
print(f.readline())
f.close()



with open('testfiles/bohr.txt', 'r') as f:
    print(f.readline())

"""

class Open:

    def __init__(self, filename, mode):
        self.filename = filename
        self.mode = mode

    def __enter__(self):
        self.f = open(self.filename, self.mode)
        return self.f

    def __exit__(self, *args):
        self.f.close()

    def readline(self):
        pass
```

## contextlib

```python
from contextlib import contextmanager

@contextmanager
def open_file(filename, mode):
    
    f = open(filename, mode)
    try:
        yield f
    finally:
        f.close()
```
